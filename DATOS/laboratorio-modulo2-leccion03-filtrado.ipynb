```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
df_union = pd.read_csv("../DATOS/union_netflix.csv")
df_union.head(2)
```

Para poder trabajar en los siguientes ejercicios, como la columna duration tiene valores en formato object, hemos añadido una columna nueva ("duracion_final") donde hemos convertido todos los datos que aparecen en duration a int. Dejando la columna duration para aquellos títulos que sean series y aparezcan como 2 Seasons, así podemos trabajar ton todos los datos y resolver todos los ejercicios.

Para ello hemos utilizado el me´todo str.extract()

```python
df_union["duracion_final"] = df_union["duration"].str.extract('(\d+)').astype(float)
```

```python
df_union.sample()
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
año_2021 = df_union["release_year"] == 2021
tipo_movie = df_union["type"] == "Movie"

movies_2021 = df_union[año_2021 & tipo_movie]
movies_2021.sample(2)
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
titulos_90min = df_union[df_union["Duration_"]== "90 min"][["title","Duration_"]]
titulos_90min
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
pais_india = df_union["country"] == "India"
despues_2018 = df_union["release_year"] > 2018

india_2018 = pd.DataFrame(df_union[pais_india & despues_2018]["title"])
india_2018
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
tv_14 = df_union["rating"] == "TV-14"
r = df_union["rating"] == "R"

tv14_r = df_union[tv_14 | r]
tv14_r
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
directores_filtro = df_union[df_union["director"].isin(['Christopher Nolan', 'Martin Scorsese'])]

conteo_titulos = directores_filtro["director"].value_counts()

print(f"Los títulos filtrados por cada uno de los directores son:")
print(conteo_titulos)
```

```python
directores_filtro
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
titulos_paises = df_union[df_union["country"].isin(["United States", "United Kingdom", "Canada"])]
titulos_paises[["title", "country","Language"]]
```

```python
#Como los títulos que tienen más de un idioma están mostrados como "Englisg/Spanisgh" interpretaremos que todos las celdas que tengan un /, entonces es que tienen más de un idioma:

varios_idiomas = titulos_paises[titulos_paises["Language"].str.contains("/", na=False)]

contar_idiomas = varios_idiomas["Language"].value_counts().sum()

print(f"El número total de títulos que pertenecen a los paises United States, United Kingdom y Canada con más de un idioma es de {contar_idiomas}")
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
años_titulos = df_union[df_union["release_year"].between(2015,2020, inclusive="both")]
años_titulos[["title","release_year"]]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
imdb = df_union["IMDB Score"].between(6.5,8.0,inclusive="both")
duracion = df_union["Duration_"].isnull()

imdb_duracion = df_union[imdb & duracion]
imdb_duracion
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
duracion90y150 = df_union[df_union["duracion_final"].between(90,150, inclusive="both")]
duracion90y150
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
titulo_love = df_union["title"].str.contains("Love", case=False, na= False)
tipo_tvshow = df_union["type"] == "TV Show"

titulo_tipo = df_union[titulo_love & tipo_tvshow]
titulo_tipo[["title", "type"]]
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
apellido_johnson = df_union[df_union["director"].str.contains("Johnson", case=False,na=False)]
print(apellido_johnson.shape)
pd.DataFrame(apellido_johnson[["title","director"]])
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
documental = df_union["listed_in"].str.contains("Documentaries", case=False, na=False)
director = df_union["director"] == "Mike Tyson"

documentales_tyson = df_union[documental & director]
documentales_tyson
```

#### Ejercicio 15: Usando `query()`, filtra todos los títulos del país 'France' con una duración de 80 minutos.

#### Ejercicio 16: Filtra los títulos que son películas ('Movie') y fueron lanzados antes de 2010.

```python
peliculas = df_union["type"] == "Movie"
año = df_union["release_year"] < 2010

peliculas_antes2010 = df_union[peliculas & año]
peliculas_antes2010[["title", "release_year"]]
```

#### Ejercicio 17: Encuentra los 10 títulos con las puntuaciones más altas en IMDb.

```python
top10_imdb = df_union.nlargest(10, "IMDB Score", keep="first")
top10_imdb[["title", "IMDB Score"]]
```

#### Ejercicio 18: Encuentra los 5 títulos con las puntuaciones más bajas en IMDb.

```python
sub10_imdb = df_union.nsmallest(10,"IMDB Score", keep="first")
sub10_imdb[["title", "IMDB Score"]]
```

#### Ejercicio 19: Filtra todos los títulos que no tienen un director asignado (valores nulos en la columna 'director').
